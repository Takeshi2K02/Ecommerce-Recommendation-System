```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def get_similar_product_details(train_data, item_name):   
    
    # Calculate cosine similarity based on the 'tags' column (product features)
    tfidf_vectorizer = TfidfVectorizer(stop_words='english')
    tfidf_matrix = tfidf_vectorizer.fit_transform(train_data['Name'])
    
    item_vector = tfidf_vectorizer.transform([item_name])

    cosine_sim = cosine_similarity(item_vector,tfidf_matrix)
    
    # Get a list of similar items (including the target item)
    similar_items = list(enumerate(cosine_sim[0]))
    
    # Sort the items by similarity score in descending order and exclude the first item (itself)
    similar_items_sorted = sorted(similar_items, key=lambda x: x[1], reverse=True)[1:16]

    # Get the indices of the similar items
    similar_items_indexed = [i[0] for i in similar_items_sorted]
    
    # Extract the details of the similar products
    similar_items_data = train_data.iloc[similar_items_indexed][['Name','Rating','Description','ImageURL','Rating_Count']]
    

    return similar_items_data
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def get_similar_product_details(train_data, item_name): 

    # Calculate cosine similarity based on the 'Name' column (product features)
    tfidf_vectorizer = TfidfVectorizer(stop_words='english')
    tfidf_matrix = tfidf_vectorizer.fit_transform(train_data['Name'])
    
    # Transform the target item
    item_vector = tfidf_vectorizer.transform([item_name])
    cosine_sim = cosine_similarity(item_vector, tfidf_matrix)
    
    # Get a list of similar items (including the target item)
    similar_items = list(enumerate(cosine_sim[0]))
    
    # Sort the items by similarity score in descending order and exclude the first item (itself)
    similar_items_sorted = sorted(similar_items, key=lambda x: x[1], reverse=True)[1:16]

    # Get the indices of the similar items
    similar_items_indexed = [i[0] for i in similar_items_sorted]
    
    # Extract the details of the similar products
    similar_items_data = train_data.iloc[similar_items_indexed][['Name', 'Rating', 'Description', 'ImageURL', 'Rating_Count']]
    
    # Assume you have a way to determine actual relevant items based on item_name
    actual_relevant_items = train_data[train_data['Name'].str.contains(item_name, case=False)]['Name'].tolist()

    # Calculate True Positives, False Positives, and False Negatives
    recommended_items = similar_items_data['Name'].tolist()

    true_positives = len(set(recommended_items) & set(actual_relevant_items))
    false_positives = len(set(recommended_items) - set(actual_relevant_items))
    false_negatives = len(set(actual_relevant_items) - set(recommended_items))

    # Calculate Precision and Recall
    precision = true_positives / (true_positives + false_positives) if (true_positives + false_positives) > 0 else 0
    recall = true_positives / (true_positives + false_negatives) if (true_positives + false_negatives) > 0 else 0

    return similar_items_data, precision, recall
```

```python
similar_products, precision, recall = get_similar_product_details(train_data, item_name='phone')
print("Similar Products:\n", similar_products)
print("Precision:", precision)
print("Recall:", recall)
```

Collaborative filtering

```python
# Function to hash and truncate user IDs
def shorten_user_id(user_id, length=8):
    return hashlib.sha256(user_id.encode()).hexdigest()[:length]

train_data['shorten_user_id'] = train_data['ID'].apply(shorten_user_id)
train_data.drop('ID', axis=1, inplace=True)
```

```python
train_data['shorten_user_id'][0]
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
import hashlib

def collaborative_filtering_recommendations(train_data, shorten_user_id, top_n=10):
    # Create the user-item matrix for ratings
    user_item_matrix = train_data.pivot_table(index='shorten_user_id', columns='ProductID', values='Rating', aggfunc='mean').fillna(0)

    # Calculate the user similarity matrix using cosine similarity
    user_similarity = cosine_similarity(user_item_matrix)

    # Find the index of the target user in the matrix
    target_user_index = user_item_matrix.index.get_loc(shorten_user_id)

    # Get the similarity scores for the target user
    user_similarities = user_similarity[target_user_index]

    # Sort the users by similarity in descending order (excluding the target user)
    similar_users_indices = user_similarities.argsort()[::-1][1:]

    # Generate recommendations based on similar users
    recommended_items = []

    for user_index in similar_users_indices:
        # Get items rated by the similar user but not by the target user
        rated_by_similar_user = user_item_matrix.iloc[user_index]
        not_rated_by_target_user = (rated_by_similar_user > 0) & (user_item_matrix.iloc[target_user_index] == 0)

        # Extract the item IDs of recommended items
        recommended_items.extend(user_item_matrix.columns[not_rated_by_target_user][:top_n])

    # Get the details of recommended items
    recommended_items_details = train_data[train_data['ProductID'].isin(recommended_items)][['Name', 'Rating_Count', 'Description', 'ImageURL', 'Rating']]
    top_items = recommended_items_details.sort_values(by='Rating', ascending=False).head(12)

    # # Calculate Precision and Recall using ratings as relevance
    # actual_interactions = train_data[train_data['shorten_user_id'] == shorten_user_id][['ProductID', 'Rating']]
    # actual_relevant_items = actual_interactions[actual_interactions['Rating'] > 0]['ProductID'].tolist()

    # # Debugging: Print relevant items
    # # print(f"Actual Relevant Items for User {shorten_user_id}: {actual_relevant_items}")

    # # True Positives: Recommended items that are also relevant
    # true_positives = [item for item in recommended_items if item in actual_relevant_items]

    # # Debugging: Print recommended items
    # # print(f"Recommended Items: {recommended_items}")

    # # False Positives: Recommended items that are not relevant
    # false_positives = [item for item in recommended_items if item not in actual_relevant_items]

    # # False Negatives: Relevant items that were not recommended
    # false_negatives = [item for item in actual_relevant_items if item not in recommended_items]

    # # Calculate Precision and Recall
    # tp = len(true_positives)
    # fp = len(false_positives)
    # fn = len(false_negatives)

    # precision = tp / (tp + fp) if (tp + fp) > 0 else 0
    # recall = tp / (tp + fn) if (tp + fn) > 0 else 0

    return recommended_items

# Example usage
# Assume you have a DataFrame called 'train_data' and a user ID
target_user_id = 'ad1c6cbe'  # Example hashed user ID
# top_items, precision, recall, actual_relevant_items, 
recommended_items = collaborative_filtering_recommendations(train_data, shorten_user_id=target_user_id, top_n=10)
print("Recommended Items:\n", top_items)
# print("Precision:", precision)
# print("Recall:", recall)
```

Hybrid Recommendations

```python
def hybrid_recommendations(train_data, shorten_user_id, item_name, top_n=10):
    # Get content-based recommendations
    content_based_rec = get_similar_product_details(train_data, item_name)

    # Get collaborative filtering recommendations
    collab_filtering_rec, _, _, _, _ = collaborative_filtering_recommendations(train_data, shorten_user_id, top_n=10)

    # Merge and deduplicate the recommendations
    hybrid_rec = pd.concat([content_based_rec, collab_filtering_rec]).drop_duplicates()
    
    return hybrid_rec.head(10)

# Example usage
hybrid_rec = hybrid_recommendations(train_data, 'ad1c6cbe', 'Phone', top_n=10)
print(hybrid_rec)
```
